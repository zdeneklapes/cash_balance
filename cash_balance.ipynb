```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cash_balance_df: pd.DataFrame = pd.DataFrame({
    'programmer_for_project': [2],
    'programmer_salary_per_month': [50_000],
    'programmer_salary_per_hour': [200],
    'month_to_develop': [2]
})
```

```python
cash_balance_df['monthly_cost_1'] = cash_balance_df['programmer_for_project'] * cash_balance_df['programmer_salary_per_month']
cash_balance_df['monthly_cost_2'] = cash_balance_df['programmer_for_project'] * cash_balance_df['programmer_salary_per_hour'] * 160
cash_balance_df['year_cost_1'] = cash_balance_df['monthly_cost_1'] * 12
cash_balance_df['year_cost_2'] = cash_balance_df['monthly_cost_2'] * 12
cash_balance_df['total_project_cost_1'] = cash_balance_df['monthly_cost_1'] * cash_balance_df['month_to_develop']
cash_balance_df['total_project_cost_2'] = cash_balance_df['monthly_cost_2'] * cash_balance_df['month_to_develop']
```

```python
cash_balance_df[['total_project_cost_1', 'total_project_cost_2']].plot(kind='bar')
```
